```python
review=sql_review_df.groupby(['reviewer_id','reviewed_item_id']).mean()
review
```

```python
review.reset_index()
```

```python
review.rating.mean()
```

```python
mean_rating=review.groupby('reviewer_id').mean()
```

```python
mean_rating=mean_rating.reset_index()
```

```python
mean_rating=mean_rating.rename(columns={'rating':'rating_offset'})
```

```python
review.reset_index(inplace=True)
```

```python
rating_normal=pd.merge(review,mean_rating,how='left',on='reviewer_id')
```

```python
rating_normal['nrating']=rating_normal['rating']-rating_normal['rating_offset']
```

```python
rating_normal
```

```python
rating_normal[rating_normal['nrating']<0]
```

```python
minR=rating_normal.nrating.min()
maxR=rating_normal.nrating.max()
```

```python
rating_normal['prating']=(rating_normal['nrating']-minR)/(maxR-minR)
```

```python
rating_normal=rating_normal[['reviewer_id','reviewed_item_id','prating']]
```

```python
rating_normal=rating_normal.rename(columns={'reviewer_id':'user_id','reviewed_item_id':'restaurant_id','prating':'rating'})
```

```python
rating_normal.to_csv('user_res_rating_df.csv',index=False)
```
