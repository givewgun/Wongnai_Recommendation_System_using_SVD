```python
res_dict=np.load('res_dict.npy').item()
user_dict=np.load('user_dict.npy').item()
res_region_dict=np.load('res_region_dict.npy').item()
user_region_dict=np.load('user_region_dict.npy').item()
user_c_region_dict=np.load('user_c_region_dict.npy').item()
df_test=pd.read_csv('test.csv',header=None, names=['user_id','number_of_res'])
```

```python
res_rank_score_dict=np.load('res_rank_score_dict.npy').item()
```

```python
rank_score_list=[1]*len(res_dict)
for i in res_dict:
    rank_score_list[res_dict[i]]=res_rank_score_dict[i]
```

```python
res_dict_inverted = {value: key for key, value in res_dict.items()}
```

```python
#all_res_factor*rank_score_list
# all_res_factor=[a*(b+5) for a,b in zip(all_res_factor,rank_score_list)]
```

```python
all_res_factor=rank_score_list
```

```python
all_res_rank=list(enumerate(all_res_factor))
all_res_rank=sorted(all_res_rank,key=lambda x:x[1],reverse=True)
all_res_rank_list=[i[0] for i in all_res_rank]
all_res_rank_list
```

```python
pd.Series(all_res_factor).describe()
```

```python
#Separate restuarant rank to each region
region_res_dict=dict()
for rid,factor in all_res_rank:
    if res_region_dict[res_dict_inverted[rid]] in region_res_dict:
        region_res_dict[res_region_dict[res_dict_inverted[rid]]].append(res_dict_inverted[rid])
    else:
        region_res_dict[res_region_dict[res_dict_inverted[rid]]]=[res_dict_inverted[rid]]
```

```python
predict=list(csv.reader(open('predict_raw.txt')))
```

```python
train_df=pd.read_csv('train.csv')
train_res_set=set(train_df.restaurant_id)
```

```python
#FILLLL
for i in range(len(df_test.values)):
    user=df_test.values[i][0]
    number=df_test.values[i][1]
    if int(predict[i][0])==0: #If 0 reset list first
        predict[i]=[]
    if number!=len(predict[i]): #If recommend list is incomplete
        if len(predict[i])!=number:
            tmp=set(predict[i])
            cregion=user_c_region_dict[user]      #Fill based on current_user_region first
            if cregion in region_res_dict:       #If c user region is avaliable in our restuarant region
                for r in region_res_dict[cregion]:
                    if(len(predict[i])==number):  #If complete then stop
                        break 
                    if True:#r in train_res_set:  #If res is in train set(redundancy)
                        if all_res_factor[res_dict[r]]<1.000000: #Limit the minimum factor score acceptable(Tunable)
                            break
                        if not r in tmp:   #Check whether the restuarant already exist in our rec list
                            predict[i].append(r)
        if len(predict[i])!=number:         #If still not complete then fill based on user_region
            tmp=set(predict[i])
            region=user_region_dict[user]
            if region in region_res_dict:
                for r in region_res_dict[region]:
                    if(len(predict[i])==number):
                        break
                    if True:#r in train_res_set:
                        if all_res_factor[res_dict[r]]<1.000000:  #Limit the minimum factor score acceptable(Tunable)
                            break
                        if not r in tmp:
                            predict[i].append(r)
        if len(predict[i])!=number:      #If still not complete then fill the remaining with restuarant with top factor of all
            tmp=set(predict[i])
            for r in all_res_rank_list:
                if(len(predict[i])==number):
                    break
                if res_dict_inverted[r] in train_res_set:
                    if not res_dict_inverted[r] in tmp:
                        predict[i].append(res_dict_inverted[r])
```

```python
with open("predict.txt","w+") as my_csv:
    csvWriter = csv.writer(my_csv,delimiter=',')
    csvWriter.writerows(predict)
```
