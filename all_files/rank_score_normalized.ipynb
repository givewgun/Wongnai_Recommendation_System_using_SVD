```python
import numpy as np
import pandas as pd
import pymysql
from sqlalchemy import create_engine
from scipy import sparse
```

```python
#must haveeee
engine = create_engine('mysql+pymysql://root:123456789@localhost:3306/wongnai')
```

```python
sql_restaurant_df = pd.read_sql_query('SELECT id,name,rank_score FROM w_restaurant', engine)
```

```python
sql_restaurant_df.sort_values(by=['rank_score'],ascending=False)
```

```python
score_range=sql_restaurant_df.rank_score.max()-sql_restaurant_df.rank_score.min()
```

```python
sql_restaurant_df['rank_score']=np.where(sql_restaurant_df.rank_score<0,0,sql_restaurant_df.rank_score)
```

```python
score_range
```

```python
sql_restaurant_df['rank_score']=sql_restaurant_df['rank_score']/score_range
```

```python
sql_restaurant_df['rank_score']=sql_restaurant_df['rank_score']+1
```

```python
sql_restaurant_df.sort_values(by=['rank_score'],ascending=False)
```

```python
res_rank_score_dict = dict(zip(sql_restaurant_df.id, sql_restaurant_df.rank_score))
```
